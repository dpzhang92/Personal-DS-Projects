## Dummy Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
%matplotlib inline
```

```python
df = pd.read_csv('dummy data.csv', index_col='id')
df = df.astype({'date':'datetime64'})
```

```python
df.describe()
```

### Group data by Hour Of Day and Day of Week

```python
hour_df = df.groupby(by=['hour_of_day']).sum()
hour_app_df = df.groupby(by=['hour_of_day','appointment']).count()

week_df = df.groupby(by=['day_of_week']).sum()
week_app_df = df.groupby(by=['day_of_week','appointment']).count()
```

```python
#Group data to calculate appointments/call
s = hour_app_df.groupby(level=0).apply(lambda x: x/x.sum())['date']
t = s.reset_index()
appointments = t[t['appointment'] == 1]
appointments.set_index('hour_of_day',inplace=True)
no_appointments = t[t['appointment'] == 0]
no_appointments.set_index('hour_of_day',inplace=True)

#Set up DF for plotting
app_df = pd.DataFrame({'Yes':appointments['date'],'No':no_appointments['date']})
```

```python
#Group data to calculate appointments/call
s = week_app_df.groupby(level=0).apply(lambda x: x/x.sum())['date']
t = s.reset_index()
appointments_week = t[t['appointment'] == 1]
appointments_week.set_index('day_of_week',inplace=True)
no_appointments_week = t[t['appointment'] == 0]
no_appointments_week.set_index('day_of_week',inplace=True)

#Set up DF for plotting
app_week_df = pd.DataFrame({'Yes':appointments_week['date'],'No':no_appointments_week['date']})
```

```python
plt.figure(figsize=(12,8))
ax1 = hour_df['appointment'].plot(kind='bar')
plt.title('Appointments By Hour Of Day')
plt.xlabel('Hour Of Day')

plt.show()
```

```python
app_df.plot.bar(rot=0, figsize=(12,8))
plt.title('Appointments/Call By Hour Of Day')
plt.xlabel('Hour Of Day')
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax1 = week_df['appointment'].plot(kind='bar')
plt.title('Appointments By Day Of Week')
plt.xlabel('Day Of Week')

plt.show()
```

```python
app_week_df.plot.bar(rot=0, figsize=(12,8))
plt.title('Appointments/Call By Day Of Week')
plt.xlabel('Day Of Week')
plt.show()
```

### Logistic Regression

```python
X = df[['hour_of_day','day_of_week']] #fit on both features
y = df[['appointment']]
clf = sm.Logit(y,X).fit() 
```

```python
print(clf.summary())
```

### Two Sample T-test

```python
#Set up DF for appointments/call
df_group = df.groupby(by=['date','appointment']).count()
r = df_group.groupby(level=0).apply(lambda x: x/x.sum())['hour_of_day']
w = r.reset_index()
samp_prev = w[w['appointment'] == 1]['hour_of_day']
```

```python
#Set up Experiment data for appointments/call
dfe = pd.read_csv('dummy data experiment.csv', index_col='id')
dfe_group = dfe.groupby(by=['date','appointment']).count()
r = dfe_group.groupby(level=0).apply(lambda x: x/x.sum())['hour_of_day']
w = r.reset_index()
samp_exp = w[w['appointment'] == 1]['hour_of_day']
```

```python
t1, p1 = stats.ttest_ind(samp_prev,samp_exp)
print('The t-statistic is {0} and the p-value is {1}'.format(t1,p1))
```

```python
abs(samp_prev.mean() - samp_exp.mean()) / samp_prev.mean()
```
